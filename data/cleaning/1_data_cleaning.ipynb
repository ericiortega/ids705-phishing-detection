### Data Cleaning of the Trec_07 dataset

```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Load the file
df = pd.read_csv(
    "https://raw.githubusercontent.com/ericiortega/ids705-phishing-detection/refs/heads/main/data/raw/trec_07.csv"
)
# Ensure upload
df
```

```python
# Combine non-null subject and body
df["clean_text"] = df["subject"].fillna("") + " " + df["body"].fillna("")
# Drop subject and body columns
df = df.drop(columns=["subject", "body"])
# Ensure the changes
df.columns
```

```python
# Convert date format
df["date"] = pd.to_datetime(df["date"], utc=True, errors="coerce")
# Check if date has redundant values
df[df["date"].dt.year > 2007]
```

```python
# Filter rows where the year is 2007
df = df[df["date"].dt.year == 2007]
# Ensure the changes
df
```

```python
# Drop redundant columns
df = df.drop(columns="date")
```

```python
# Reorder
df = df[["sender", "receiver", "clean_text", "urls", "label"]]
# Final clean data
df
```

```python
# Save DataFrame to CSV
df.to_csv("2_cleaned_data.csv", index=False)
```
